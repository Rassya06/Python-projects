```python
!pip install folium
import pandas as pd
import folium
import plotly.graph_objects as go
cities = pd.read_csv("katalog_gempa.csv")
print(cities)
```

```python
cities.info()
```

```python
cities=cities.drop(['strike1','dip1','rake1','strike2','dip2','rake2'],axis=1)
cities
```

```python
strong_quake=cities[cities['mag']>=6.1]
```

```python
strong_quake.info()
```

```python
strong_quake.drop_duplicates(inplace=True)
strong_quake
```

```python
strong_quake.duplicated()
```

```python
strong_quake.isna().any()
```

```python
print(strong_quake.remark.unique())
```

```python
import numpy as np
earthquake = strong_quake.copy()
earthquake['category'] = np.select(
    [(earthquake['mag'].between(6.0,7.9,inclusive='right')),
     (earthquake['mag'].between(7.9,8,inclusive='right')),
     (earthquake['mag']>8)],
    ['Strong', 'Major','Great'],
    default='Undetermined'
)
earthquake
```

```python
#map of major earthquakes
fig = go.Figure(data=go.Scattergeo(
    lat=strong_quake.lat,
    lon=strong_quake.lon,
    text=strong_quake.remark,
    marker=dict(
        color=strong_quake.mag,
        colorscale='burg',
        opacity=1,
        size=strong_quake['mag']**1.3,
        colorbar=dict(
            titleside="right",
            outlinecolor="rgba(0,68,128,255)",
            ticks="outside",
            showticksuffix="last"
        )
    )
))
fig.update_layout(
    geo = dict(
        scope = 'asia',
        showland = True,
        subunitcolor = "rgb(255, 128, 255)",
        countrycolor = 'grey',
        showsubunits = True,
        showcountries = True,
        resolution = 50,
        lonaxis = dict(
            showgrid = True,
            gridwidth = 0.5,
            range= [ 94.0, 142.0 ],
            dtick = 5
        ),
        lataxis = dict (
            showgrid = True,
            gridwidth = 0.5,
            range= [ -6.0, 11.0 ],
            dtick = 5
        )
    ),
    title='Map of major earthquakes in Indonesia',
)
fig.show()
```
